#Question 5

```python
import pandas as pd

data = {'Age': [25, 30, 150, -5, 45, 60, 10, 130, 70, 35]}
df = pd.DataFrame(data)
display(df)

outlier_mask = (df['Age'] < 0) | (df['Age'] > 120)
outliers = df[outlier_mask]
display(outliers)

df_cleaned = df[~outlier_mask].copy()
display(df_cleaned)

print("Dataset before outlier removal:")
display(df)

print("\nDataset after outlier removal:")
display(df_cleaned)
```

#Question 6

```python
data = {'Income': [50000, 65000, 30000, 120000, 80000, 45000, 90000, 75000, 110000, 55000]}
df_income = pd.DataFrame(data)
display(df_income)

income_min = df_income['Income'].min()
income_max = df_income['Income'].max()
df_income['Income_Normalized'] = (df_income['Income'] - income_min) / (income_max - income_min)
display(df_income)

income_mean = df_income['Income'].mean()
income_std = df_income['Income'].std()
df_income['Income_Standardized'] = (df_income['Income'] - income_mean) / income_std
display(df_income)

display(df_income)
```
